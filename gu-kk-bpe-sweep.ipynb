```python
import os
import re
```

```python
import pandas as pd
```

```python
results_path = 'evaluate/kk_gu_bpe_sweep/'
```

```python
bpe_sweep_csv_names = os.listdir(results_path)
```

```python
def parse_filename(fname):
    lang, _, bpe_csv = fname.split('_')
    bpe = int(bpe_csv.replace('bpe', '')
                     .replace('.csv', ''))
    return (lang, bpe)
```

```python
all_results = pd.DataFrame()
```

```python
bpes = [1000, 2500, 1000, 7500, 10000]
```

```python
for fname in bpe_sweep_csv_names:
    lang, bpe_size = parse_filename(fname)
    df = pd.read_csv(f"{results_path}/{fname}")
    df['bpe_size'] = bpe_size
    df['method'] = df.method.replace('baseline', 'sentencepiece')
    all_results = pd.concat([all_results, df])
```

```python
all_results['method'] = all_results.method.str.replace('baseline', 'sentencepiece')
```

```python
(all_results.groupby(["pair", "method", "bpe_size"]).bleu.describe().round(3))[['mean', 'std', '50%']]#.plot(kind='boxplot')
```

```python
(all_results.groupby(["pair", "method", "bpe_size"]).chrf3.describe().round(3)*100)[['mean', 'std', '50%']]#.plot(kind='boxplot')
```
